```python
import sys
import os

project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))

if project_root not in sys.path:
    sys.path.append(project_root)

from engines.yolo_npu_engine import YoloNpuEngine
import cv2
```

```python
engine = YoloNpuEngine(model_path="../yolo_models/hand-quantized.onnx")
```

```python
cap = cv2.VideoCapture(0)
```

```python
## Loop to read and display frames

while True:
    ret, frame = cap.read() #
    
    if not ret:
        break #

    results = engine.infer(frame)
    annotated_image = engine.draw_results(frame, results)

    

    # Display the frame
    cv2.imshow('Live Video Feed', annotated_image) #

    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Cleanup
cap.release() #
cv2.destroyAllWindows() #
```
