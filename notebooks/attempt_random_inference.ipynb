```python
import onnxruntime as ort
import numpy as np
```

```python
#Generate dummy image
dummy_image = np.random.rand(1,3,224,224).astype(np.float32)
```

```python
#Load model
MODEL_PATH = "../yolo_models/hand-trained.onnx"
```

```python
#Run inference
try:
    # Loading on CPU just to test the tensor shapes quickly
    session = ort.InferenceSession(MODEL_PATH, providers=['CPUExecutionProvider'])
    input_name = session.get_inputs()[0].name
    
    # 3. Run Inference
    print("\nRunning inference...")
    outputs = session.run(None, {input_name: dummy_image})
    raw_tensor = outputs[0] 
    
except Exception as e:
    print(f"\nModel not found at '{MODEL_PATH}'.")
    print("Simulating the [1, 68, 1029] NPU output tensor to demonstrate the math...")
    # Simulating the exact shape the NPU would output
    raw_tensor = np.random.rand(1, 68, 1029).astype(np.float32)
```
