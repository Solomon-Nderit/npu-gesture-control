```python
import math
import numpy as np
```

```python
#Function to calculate the distance between two points passed in as tuples

def calculate_dist(point_1, point_2):
    x1 = point_1[0]
    x2 = point_2[0]

    y1 = point_1[1]
    y2 = point_2[1]

    dist = math.sqrt(((x2-x1)**2)+((y2-y1)**2))

    return dist
```

```python
calculate_dist((0.2,0.4),(0.6,0.7))
```

```python
#S to return various gestures and actions depending on the coordinates of various points

def click_detector(hand_landmarks):
    thumb_tip = hand_landmarks["thumb_tip"]
    index_tip = hand_landmarks["index_tip"]
    dist = calculate_dist(thumb_tip, index_tip)

    if dist<=0.2:
        return "click"

    return "no click"
```

```python
hand_landmarks = {
    "thumb_tip":(0.2,0.7),
    "index_tip": (0.2,0.8)
}

click_detector(hand_landmarks)

```

```python
#Generate random coordinates and detect a click
# random_array = np.random.rand(2)
# print(random_array)

start = 0.0
stop = 1.0

for i in range(20):
    rnge = np.random.default_rng()
    random_thumb_coord = rnge.uniform(start, stop, size=2)

    random_index_coord = rnge.uniform(start, stop, size=2)

    hand_landmarks = {
        "thumb_tip":random_thumb_coord,
        "index_tip":random_index_coord
    }
    
    print(click_detector(hand_landmarks))

```
