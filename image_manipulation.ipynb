```python
import pandas as pandas
import numpy as numpy

from glob import glob


import cv2
import matplotlib.pylab as plt
```

### Reading in Images

```python
dog_files=glob('./cat-dog-dataset/training_set/training_set/dogs/*.jpg')
cat_files=glob('./cat-dog-dataset/training_set/training_set/cats/*.jpg')
```

```python
#Reading and importing an image using matplotlib
img_mpl = plt.imread(cat_files[19])

#Reading and importing an image using cv2
img_cv2=cv2.imread(cat_files[20])
```

### Display Images

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(img_mpl)
plt.show()
```

```python
# Displaying RGB Channels of our image
fig, axs = plt.subplots(1,3, figsize=(15,5))
axs[0].imshow(img_mpl[:,:,0], cmap='Reds')
axs[1].imshow(img_mpl[:,:,1], cmap='Greens')
axs[2].imshow(img_mpl[:,:,2], cmap='Blues')

```
