## Crop an image by modifying it as a numpy array.

```python
#Function to crop an image into a square of 200 by 200 pixels by finding the middle 200 pixels in both rows and columns
def crop_image(image, d_height, d_width):
    height, width, channels = image.shape

    start_height = (height//2) - (d_height//2)
    end_height = start_height + d_height

    start_width = (width//2) - (d_width//2)
    end_width = start_width+d_width

    new_image = image[start_height:end_height, start_width:end_width, :]

    return new_image

```

```python
my_image = crop_image(image, 200, 200)

cv2.imwrite('images/cropped_image.jpg',my_image)

my_image.shape
```

## Transposing the shape of a Numpy array

```python
# 1. Transposing by using the .T attribute which automatically flips the axes
image.shape

T_transposed_image = image.T

T_transposed_image.shape
```

```python
# 2. Transposing by using the .transpose() function and manually specifying the new axes
image.shape

man_transposed_image = np.transpose(image, (2, 1, 0))

man_transposed_image.shape
```

## Adding a new dimension to a numpy array


### Using np.newaxis


```python
image.shape
```

```python
#Dimension at the beginning
new_image = image[np.newaxis, :]

new_image.shape
```

```python
#Dimension at the middle

new_image_2 = image[:, np.newaxis, :, :]
new_image_2.shape
```

```python
#Dimension at the end
new_image_3 = image[..., np.newaxis]
new_image_3.shape

```

### Using np.expand_dims()


```python
#Dimension at the beginning
image.shape
```

```python
expanded_image = np.expand_dims(image, axis = 0)
expanded_image.shape
```

```python
#Dimension at end
expanded_image_2 = np.expand_dims(image, axis=-1)
expanded_image_2.shape
```

## Numpy integer array conversion to float

```python
float_image = image.astype(float)

float_image.shape
```

```python
# Accessing a single pixel to understand how the data is stored
unfortunate_pixel = image[100,100]
print(unfortunate_pixel)
```

```python
#In the above case, since OpenCV stores images in BGR, the pixel has a strength of 254 blue, 249 green and 234 red

#If we wanted to acess the blue value of a particular pixel, we can specify the channel

unfortunate_blue_of_pixel = image[100,100,0]

print(unfortunate_blue_of_pixel)
```

```python
# To convert to float (i.e. normalization), we convert the data type to float32 then divide by the highest possible value i.e. 255
float_image = image.astype(np.float32)
image_normalized = float_image/255.0
```

```python
#Let's now locate our original unfortunate pixel and witness his mutation
mutated_pixel = image_normalized[100,100]
print(mutated_pixel)
```
