```python
import onnxruntime as ort
import numpy as np
```

```python
session = ort.InferenceSession("models/test_model.onnx", providers=["CPUExecutionProvider"])
```

## Find the shape of data the model accepts


```python
#Input metadata
input_metadata = session.get_inputs()[0]
input_name = input_metadata.name
input_shape = input_metadata.shape
input_type = input_metadata.type
```

## Create some dummy data that matches the input data

```python
numpy_type = np.float32
```

```python
dummy_shape = [1 if dim is None or isinstance(dim, str) else dim for dim in input_shape]

dummy_input = np.random.rand(*dummy_shape).astype(numpy_type)
```

## Run inference


```python
outputs = session.run(None, {input_name:dummy_input})
```
